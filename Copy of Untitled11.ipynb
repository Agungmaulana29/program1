```python
# Membuat fungsi dengan parameter
def luas_segitiga(alas, tinggi):
    luas = (alas * tinggi) / 2
    print ("Luas segitiga: %f" % luas)

# Pemanggilan fungsi
luas_segitiga(4, 6)
```

```python
def luas_persegi(sisi):
    luas = sisi * sisi
    return luas

# Pemanggilan fungsi
print ("Luas persegi: %d" % luas_persegi(6))
```

```python
# membuat variabel global
nama = "Belajar Kode"
versi = "1.0.0"

def help():
  # ini variabel lokal
  nama = "Programku"
  versi = "1.0.1"
  # mengakses variabel lokal
  print ("Nama: %s" % nama)
  print ("versi: %s" % versi)

# mengakses variabel global
print ("Nama: %s" % nama)
print ("versi: %s" % versi)

# memanggil fungsi help()
help()
```

```python
# membuat variabel global
nama = "agung maulana"
tanggal_lahir = "29-maret-2004"

def help():
  # ini variabel lokal
  hobi = "bermain game"
  jurusan = "ILKOM"
  tempat_tinggal = "desa pangkal buluh"
  # mengakses variabel lokal
  print ("hobi: %s" % hobi)
  print ("jurusan: %s" % jurusan)
  print ("tempat_tinggal: %s" % tempat_tinggal)

# mengakses variabel global
print ("Nama: %s" % nama)
print ("tanggal_lahir: %s" % tanggal_lahir)

# memanggil fungsi help()
help()
```

```python
buku = []
# fungsi untuk menampilkan semua data
def show_data():
    if len(buku) <= 0:
        print("BELUM ADA DATA")
    else:
        for indeks in range(len(buku)):
            print("[{}] {}".format(indeks, buku[indeks]))

# fungsi untuk menambah data
def insert_data():
    buku_baru = input("Judul Buku:")
    buku.append(buku_baru)

# fungsi untuk edit data
def edit_data():
    show_data()
    indeks = int(input("Inputan ID buku"))
    if indeks > len(buku):
      print("ID salah")
    else:
      judul_baru = input("Judul Buku:")
      buku[indeks] = judul_baru

def delete_data():
    show_data()
    indeks = int(input("Inputan ID buku"))
    if indeks > len(buku):
      print("ID salah")
    else:
      buku.remove(buku[indeks])

# fungsi untuk menampilkan menu
def show_menu():
    print("\n")
    print("---------- MENU ----------")
    print("[1] Show Data")
    print("[2] Insert Data")
    print("[3] Edit Data")
    print("[4] Delete Data12345678")
    print("[5] Exit")

    menu = input("PILIH MENU>")
    print("\n")

    if int(menu) == 1:
      show_data()
    elif int(menu) == 2:
      insert_data()
    elif int(menu) == 3:
      edit_data()
    elif int(menu) == 4:
      delete_data()
    elif int(menu) == 5:
     exit()
    else:
      print("Salah pilih!")

if __name__ == "__main__":

    while True:
        show_menu()
```
